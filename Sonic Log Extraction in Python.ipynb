Python has a well developed library LASIO for reading the well log .las files

```python
import lasio as la
import pandas as pd
```

```python
las = la.read(r'4200341370.las')
```

View the log description

```python
print(las.curves)
```

Import the data frame

```python
las_df = pd.DataFrame(las.data, columns=list(las.curvesdict.keys()))
```

```python
las_df.head()
```

```python
print(las_df.info())
print(las_df.shape)
```

```python
las_df.describe()
```

```python
any(las_df.isna())
```

```python
#Drop missing values - sometimes the calibration tools malfunction and do not read the values
las_df.dropna(axis=0,how='any',inplace=True)
las_df.shape
```

```python
las_df.head()
```

```python
las_df.to_csv("las_df.csv")
```
